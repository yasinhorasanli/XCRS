```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

def plot_bar_chart(labels, sizes, x_y=["Ratings", "Number of Participants"]):

    #colors = ['royalblue', 'seagreen', 'darkorange', 'firebrick', 'darkorchid'] 
    colors = cm.Blues(np.linspace(0.3, 1, len(labels)))

    plt.bar(labels, sizes, color=colors[:len(labels)])

    # plt.title('Column Chart')
    plt.xlabel(x_y[0])
    plt.ylabel(x_y[1])

    # plt.xticks(rotation=30, ha="right")
    max_value = max([max(sizes)])
    plt.yticks(np.arange(1, max_value+1, 2)) 

    # Adjust the layout to make room for the rotated labels
    plt.tight_layout()

    # Display the chart
    plt.show()

def plot_pie_chart(labels, sizes):

    colors = ['royalblue', 'seagreen', 'darkorange', 'firebrick', 'darkorchid'] 

    def autopct_func(pct):
        return ('%1.1f%%' % pct) if pct > 0 else ''

    # Plotting the pie chart without labels
    plt.pie(sizes, colors=colors[:len(labels)], autopct=autopct_func, shadow=False, startangle=90, counterclock=False)

    # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.axis('equal')

    # Adding a legend at the bottom
    #plt.legend(labels, loc="lower center", bbox_to_anchor=(0.5, -0.1), ncol=len(labels))
    plt.legend(labels, loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol=len(labels))
    
    # Title of the chart
    # plt.title('Pie Chart with Labels at the Bottom')

    # Display the chart
    plt.show()
    
```

# Effectiveness

How satisfied are you with the relevance of the role and course recommendations? (Effectiveness)	\

Average: 4 80%

```python

labels = ["1", "2", "3", "4", "5"]
sizes = [0,1,6,10,8]  


plot_bar_chart(labels=labels, sizes=sizes)
```
