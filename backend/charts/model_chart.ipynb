```python
import matplotlib.pyplot as plt
import numpy as np

def plot_bar_chart(sizes, x_y):
    # Model names and their corresponding colors
    model_names = ["text-embedding-004 \n (Google)", "voyage-large-2 \n (VoyageAI)", "text-embedding-3-large \n (OpenAI)", "mistral-embed \n (MistralAI)", "embed-english-v3.0 \n (Cohere)"]
    colors = ['royalblue', 'seagreen', 'darkorange', 'firebrick', 'darkorchid']

    # X-axis labels will be model numbers (e.g., "Model 1", "Model 2", etc.)
    labels = [f"Model {i+1}" for i in range(len(sizes))]

    plt.bar(labels, sizes, color=colors[:len(labels)])

    # Adding a legend by creating a set of invisible bars for each color and label
    handles = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(model_names))]
    plt.legend(handles, model_names, title="Models", fontsize='x-small', title_fontsize='medium', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.xlabel(x_y[0])
    plt.ylabel(x_y[1])

    max_value = max([max(sizes)])
    plt.yticks(np.arange(0, max_value+2, 1)) 

    #plt.xticks(rotation=30, ha="right")

    # Adjust the layout to make room for the rotated labels
    plt.tight_layout()

    # Display the chart
    plt.show()


```

```python
from scipy.stats import friedmanchisquare

# Career roles responses for Models 1-5
model_1_career_roles = [5, 5, 5, 3, 4, 3, 3, 4, 3, 2, 1, 2, 4, 5, 3, 2, 3, 2, 5, 3, 2, 5, 4, 5, 5]
model_2_career_roles = [4, 5, 5, 4, 4, 2, 5, 5, 3, 2, 1, 5, 5, 4, 4, 5, 2, 2, 3, 3, 3, 4, 3, 3, 5]
model_3_career_roles = [3, 4, 3, 3, 4, 3, 3, 4, 5, 2, 3, 4, 3, 3, 3, 3, 3, 3, 3, 1, 4, 3, 5, 4, 4]
model_4_career_roles = [2, 4, 4, 4, 4, 5, 4, 4, 4, 4, 3, 4, 4, 4, 2, 5, 3, 2, 3, 2, 4, 2, 4, 4, 5]
model_5_career_roles = [3, 4, 5, 5, 3, 4, 1, 5, 4, 5, 4, 5, 2, 3, 2, 1, 2, 1, 2, 5, 5, 5, 4, 4, 5]

# Courses responses for Models 1-5
model_1_courses = [4, 5, 4, 4, 3, 3, 4, 5, 3, 3, 1, 2, 2, 4, 2, 2, 3, 3, 5, 4, 2, 2, 5, 5, 5]
model_2_courses = [3, 5, 4, 5, 2, 2, 5, 3, 3, 3, 1, 5, 4, 4, 2, 4, 2, 3, 4, 3, 2, 2, 4, 3, 5]
model_3_courses = [4, 3, 4, 4, 3, 3, 3, 3, 5, 3, 3, 4, 3, 4, 2, 3, 2, 3, 4, 3, 3, 2, 5, 5, 4]
model_4_courses = [2, 4, 3, 4, 4, 5, 4, 2, 4, 4, 3, 4, 4, 3, 4, 5, 3, 4, 3, 3, 4, 2, 5, 5, 5]
model_5_courses = [2, 4, 4, 4, 3, 4, 1, 4, 4, 5, 4, 5, 2, 4, 3, 2, 4, 2, 4, 4, 4, 3, 5, 3, 4]

# Perform Friedman test for career roles
result_career_roles = friedmanchisquare(
    model_1_career_roles, model_2_career_roles, model_3_career_roles, model_4_career_roles, model_5_career_roles
)

# Perform Friedman test for courses
result_courses = friedmanchisquare(
    model_1_courses, model_2_courses, model_3_courses, model_4_courses, model_5_courses
)

print("Friedman Test Result for Career Roles:", result_career_roles)
print("Friedman Test Result for Courses:", result_courses)
```
