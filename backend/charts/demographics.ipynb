```python
import matplotlib.pyplot as plt


def plot_pie_chart(labels, sizes):

    colors = ['royalblue', 'seagreen', 'darkorange', 'firebrick'] 

    # Plotting the pie chart without labels
    plt.pie(sizes, colors=colors[:len(labels)], autopct='%1.1f%%', shadow=False, startangle=90, counterclock=False)

    # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.axis('equal')

    # Adding a legend at the bottom
    #plt.legend(labels, loc="lower center", bbox_to_anchor=(0.5, -0.1), ncol=len(labels))
    plt.legend(labels, loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol=2)
    
    # Title of the chart
    # plt.title('Pie Chart with Labels at the Bottom')

    # Display the chart
    plt.show()
    
```

# Education Status

12 Bachelor's degree \
7 Currently enrolled in a Bachelor's degree program \
2 Currently enrolled in a Master's degree program \
4 Master's degree

```python

labels = ["Bachelor's degree student", "Bachelor's degree", "Master's degree student", "Master's degree"]
sizes = [7,12,2,4]  

plot_pie_chart(labels=labels, sizes=sizes)
```
