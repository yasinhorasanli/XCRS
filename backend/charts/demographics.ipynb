```python
import matplotlib.pyplot as plt


def plot_bar_chart(labels, sizes, x_y):

    colors = ['royalblue', 'seagreen', 'darkorange', 'firebrick']  

    plt.bar(labels, sizes, color=colors[:len(labels)])

    # plt.title('Column Chart')

    plt.xlabel(x_y[0])
    plt.ylabel(x_y[1])

    #plt.xticks(rotation=15, ha="right")

    # Adjust the layout to make room for the rotated labels
    plt.tight_layout()

    # Display the chart
    plt.show()
```

# Education Status

12 Bachelor's degree \
7 Currently enrolled in a Bachelor's degree program \
2 Currently enrolled in a Master's degree program \
4 Master's degree

```python

labels = ["Bachelor Student", "Bachelor", "Master Student", "Master"]
sizes = [7,12,2,4]  
x_y = ["Education Level", "Number of Participants"]

plot_bar_chart(labels=labels, sizes=sizes, x_y=x_y)
```
