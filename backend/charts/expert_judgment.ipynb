```python
import matplotlib.pyplot as plt

def plot_stacked_bar_chart(personaA, personaB, personaC):
    labels = ["1", "2", "3", "4", "5"]

    # Create a stacked bar chart
    plt.bar(labels, personaA, color='salmon', label='Persona A')
    plt.bar(labels, personaB, bottom=personaA, color='steelblue', label='Persona B')
    plt.bar(labels, personaC, bottom=[i+j for i,j in zip(personaA, personaB)], color='darkgrey', label='Persona C')

    plt.xlabel('Categories')
    plt.ylabel('Values')
    #plt.title('Stacked Bar Chart with Three Subcategories')
    plt.legend()

    # Set y-axis to display only integer values
    max_value = max([i+j+k for i,j,k in zip(personaA, personaB, personaC)])
    plt.yticks(range(0, max_value + 1))

    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm


def plot_grouped_bar_chart(roles_scores, courses_scores):
    labels = ["1", "2", "3", "4", "5"]
    x = np.arange(len(labels))  # the label locations
    width = 0.35  # the width of the bars

    # Create a grouped bar chart
    plt.bar(x - width/2, roles_scores, width, color='darkorchid', label='Role Recommendation')
    plt.bar(x + width/2, courses_scores, width, color='darkorange', label='Course Recommendation')

    plt.xlabel('Ratings')
    plt.ylabel('Number of Selections')
    plt.xticks(x, labels)  # Set the labels for x-axis

    # Set y-axis to display only integer values
    max_value = max([max(roles_scores), max(courses_scores)])
    plt.yticks(range(0, max_value + 1))

    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_bar_chart(labels, sizes, x_y=["Ratings", "Number of Selections"]):

    # colors = ['darkorchid', 'darkorange', 'seagreen', 'firebrick', 'royalblue'] 

    colors = cm.Reds(np.linspace(0.3, 1, len(labels)))
    plt.bar(labels, sizes, color=colors)

    # plt.title('Column Chart')
    plt.xlabel(x_y[0])
    plt.ylabel(x_y[1])

    # plt.xticks(rotation=30, ha="right")

    # Adjust the layout to make room for the rotated labels
    plt.tight_layout()

    # Display the chart
    plt.show()
```
